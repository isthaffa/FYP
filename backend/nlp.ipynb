```python
import regex as re
def file_to_sentence_list(file_path):
	with open(file_path, 'r',encoding = "utf8") as file:
		text = file.read()

	# Splitting the text into sentences using
	# delimiters like '.', '?', and '!'
	sentences = [sentence.strip() for sentence in re.split(
		r'(?<=[.!?])\s+', text) if sentence.strip()]

	return sentences

file_path = 'words.txt'
sentences = file_to_sentence_list(file_path)
```

```python
# Tokenization of each document
tokenized_sent = []
for s in sentences:
    tokenized_sent.append(word_tokenize(s.lower()))
tokenized_sent
```

```python
def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
```

```python
tagged_data = [TaggedDocument(d, [i]) for i, d in enumerate(tokenized_sent)]
tagged_data
```

```python
model = Doc2Vec(tagged_data, vector_size = 20, window = 2, min_count = 1, epochs = 5000)
model.save("d2v.model")
model.wv.index_to_key
```

```python
test_doc = word_tokenize("ගණිතය මම කැමතියි ")
test_doc_vector = model.infer_vector(test_doc)
model.docvecs.most_similar(positive = [test_doc_vector])
```

```python
import os

def sentence_prediction(words_input):
    model = Doc2Vec.load(
        os.path.join('d2v.model'))
    predicted_words = word_tokenize(words_input.lower())
    predicted_sentence = model.infer_vector(predicted_words)
    highest_similarity = model.docvecs.most_similar(positive=[predicted_sentence])
    return highest_similarity[0][0]
```

```python
tag = sentence_prediction("දුයි")
output = sentences[tag]

print(f"f{output}")
```
